```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/customer_segmentation_data.csv")
df.head()
```

# Visualisasi Demografi Pelanggan

```python
# Histogram (Matplotlib): Distribusi Usia Pelanggan
plt.figure(figsize=(8, 5))
plt.hist(df['age'], bins=20, color='pink', edgecolor='black')
plt.title('Distribusi Usia Pelanggan')
plt.xlabel('Usia (Tahun)')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Pie Chart (Matplotlib): Distribusi Jenis Kelamin Pelanggan
plt.figure(figsize=(6, 6))
plt.pie(df['gender'].value_counts(), labels=df['gender'].value_counts().index, autopct='%1.1f%%', colors=['lightcoral', 'lightblue', 'lightpink', 'gold'])
plt.title('Distribusi Jenis Kelamin Pelanggan')
plt.show()
```

```python
# Histogram (Matplotlib): Distribusi Pendapatan Pelanggan
plt.figure(figsize=(10, 6))
plt.hist(df['spending_score'], bins=30, color='pink', edgecolor='black')
plt.title('Distribusi Pendapatan Pelanggan')
plt.xlabel('Pendapatan')
plt.ylabel('Frekuensi')
plt.show()
```

# Visualisasi Perilaku Pembelian


```python
# Scatter Plot (Matplotlib): Hubungan antara Pendapatan dan Skor Pengeluaran
plt.figure(figsize=(8, 5))
plt.scatter(df['spending_score'], df['income'], c='red', alpha=0.5)
plt.title('Hubungan antara Pendapatan dan Skor Pengeluaran')
plt.xlabel('Spending Score')
plt.ylabel('Income')
plt.show()
```

```python
# Scatter Plot (Matplotlib): hubungan antara frekuensi pembelian dan skor pengeluaran
plt.figure(figsize=(8, 5))
plt.scatter(df['purchase_frequency'], df['spending_score'], c='red', alpha=0.5)
plt.title('Hubungan antara Pendapatan dan Skor Pengeluaran')
plt.xlabel('purchase frequency')
plt.ylabel('spending score')
plt.show()
```

```python
# Bar Chart (Seaborn) Kategori Produk yang Disukai
plt.figure(figsize=(8, 5))
sns.countplot(x='preferred_category', data=df, palette='viridis')
plt.title('Distribusi Kategori Produk yang Disukai')
plt.xlabel('Kategori Produk')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=45)
plt.show()
```

# Visualisasi Loyalitas Pelanggan

```python
# Scatter Plot (Matplotlib): Hubungan antara lama keanggotaan dan skor pengeluaran.
plt.figure(figsize=(8, 5))
plt.scatter(df['membership_years'], df['spending_score'], c='pink', alpha=0.5)
plt.title('Spending Score vs. Membership Years')
plt.xlabel('Membership Years')
plt.ylabel('Spending Score')
plt.show()
```

```python
# Histogram (Matplotlib): Distribusi Lama Keanggotaan
plt.figure(figsize=(10, 6))
plt.hist(df['membership_years'], bins=30, color='pink', edgecolor='black')
plt.title('Distribusi Lama Keanggotaan Pelanggan')
plt.xlabel('Lama Keanggotaan (Tahun)')
plt.ylabel('Frekuensi')
plt.show()
```

# Visualisasi Tren Pembelian

```python
# Scatter Plot (Seaborn): Jumlah Pembelian Terakhir vs. Spending Score
plt.figure(figsize=(8, 5))
sns.scatterplot(x='last_purchase_amount', y='spending_score', hue='gender', data=df)
plt.title('Spending Score vs. Last Purchase Amount')
plt.xlabel('Last Purchase Amount')
plt.ylabel('Spending Score')
plt.legend(title='Gender')
plt.show()
```

```python
#Bar Chart (Seaborn): Distribusi Kategori Produk Berdasarkan Segmen
df['segment'] = pd.cut(df['age'], bins=[0, 30, 60, 100], labels=['Muda', 'Dewasa', 'Senior'])
contingency_table = pd.crosstab(df['segment'], df['preferred_category'])

contingency_table.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Distribusi Kategori Produk Berdasarkan Segmen Usia')
plt.xlabel('Segmen Usia')
plt.ylabel('Jumlah Pelanggan')
plt.xticks(rotation=0)
plt.legend(title='Kategori Produk')
plt.tight_layout()
plt.show()
```
